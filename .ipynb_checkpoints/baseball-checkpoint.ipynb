```python
pitcher_names = {
# 04012026
# ARI - Ryne Nelson
# KC - Cole Ragans
# SF - Robbie Ray    
    "Nelson": "Ryne",
    "Ragans": "Cole",
    "Ray": "Robbie"

}
```

```python
batter_results = {}

for last, first in batter_names.items():
    # playerid_lookup handles unicode, accents, and Jr. suffixes
    df = playerid_lookup(last, first)

    if df.empty:
        batter_results[f"{first} {last}"] = None
    else:
        # Take the first match (usually correct)
        batter_results[f"{first} {last}"] = int(df.iloc[0]["key_mlbam"])

# Convert to a DataFrame if useful:
id_df = pd.DataFrame.from_dict(batter_results, orient="index", columns=["mlbam_id"])
print(id_df)
```

```python
from pybaseball import playerid_lookup, statcast_pitcher
import pandas as pd

def get_pvb(batter_first, batter_last, pitcher_first, pitcher_last,
            start="2010-01-01", end="2026-03-31"):
    """
    Returns:
      - pvb_data: full pitch-by-pitch Statcast DataFrame for the matchup
      - summary: dictionary of aggregate matchup stats, now including RBI, doubles, triples
    """

    # 1. MLBAM ID lookup
    batter_id = playerid_lookup(batter_last, batter_first).iloc[0]["key_mlbam"]
    pitcher_id = playerid_lookup(pitcher_last, pitcher_first).iloc[0]["key_mlbam"]

    # 2. Pull pitcher Statcast data
    p_data = statcast_pitcher(start, end, pitcher_id)

    # 3. Filter for batter
    pvb = p_data[p_data["batter"] == batter_id]

    # 4. Build summary stats
    ab_events = ["single", "double", "triple", "home_run", "field_out"]

    summary = {
        "PA": len(pvb.dropna(subset=["events"])),
        "AB": sum(pvb["events"].isin(ab_events)),
        "H": sum(pvb["events"].isin(["single", "double", "triple", "home_run"])),
        "1B": sum(pvb["events"] == "single"),
        "2B": sum(pvb["events"] == "double"),
        "3B": sum(pvb["events"] == "triple"),
        "HR": sum(pvb["events"] == "home_run"),
        "RBI": pvb["rbi"].sum() if "rbi" in pvb.columns else 0,
        "BB": sum(pvb["events"] == "walk"),
        "K": sum(pvb["events"] == "strikeout"),
    }

    return summary
```

```python
todays_bat = [get_pvb("Austin","Riley","Ryne","Nelson"),
get_pvb("Luis","Robert","Robbie","Ray"),
get_pvb("Matt","Wallner","Cole","Ragans")]
```
